```python
import os
from pathlib import Path

# Setup path
activity_path = Path.home() / "Documents" / "Barte_Activity_5"
files = [f for f in activity_path.glob("*.txt")]
stats = []

for file_path in files:
    with open(file_path, "r", encoding="utf-8") as f:
        lines = [line for line in f if line.strip()] # Ignore empty lines
        text = " ".join(lines)
        words = text.split()
        
        line_count = len(lines)
        word_count = len(words)
        char_count = len(text)
        
        # Calculations
        wpl = word_count / line_count if line_count > 0 else 0
        cpw = char_count / word_count if word_count > 0 else 0
        
        stats.append({
            "name": file_path.name,
            "wpl": wpl,
            "cpw": cpw,
            "density": word_count / char_count if char_count > 0 else 0
        })

# Sort by word density (highest to lowest)
stats.sort(key=lambda x: x['density'], reverse=True)

print(f"{'Filename':<25} | {'Words/Line':<10} | {'Chars/Word':<10}")
print("-" * 50)
for s in stats:
    print(f"{s['name']:<25} | {s['wpl']:<10.2f} | {s['cpw']:<10.2f}")
```

```python
import collections
import string

stopwords = ["the", "is", "and", "to", "in", "it", "of", "a", "with", "for"]
all_words = []

for file_path in activity_path.glob("*.txt"):
    with open(file_path, "r", encoding="utf-8") as f:
        # Normalize: lowercase and remove punctuation
        content = f.read().lower()
        content = content.translate(str.maketrans('', '', string.punctuation))
        
        # Filter stopwords
        words = [w for w in content.split() if w not in stopwords]
        all_words.extend(words)

# Identify trends
word_counts = collections.Counter(all_words)
print("Top 5 Most Frequent Meaningful Words:")
for word, count in word_counts.most_common(5):
    print(f"{word}: {count}")
```

```python
import shutil
from datetime import datetime

# Student Info
student_id = "2025-0419"
student_name = "Daniel Barte"

def smart_backup(target_file):
    target = Path(target_file)
    if not target.exists(): return
    
    # Create backup directory
    backup_dir = target.parent / f"backup_{student_id}"
    backup_dir.mkdir(exist_ok=True)
    
    # Construct filename: {original}_{id}_{timestamp}.txt
    timestamp = datetime.now().strftime("%Y%m%d_%H%M%S")
    backup_name = f"{target.stem}_{student_id}_{timestamp}{target.suffix}"
    backup_path = backup_dir / backup_name
    
    # Copy file
    shutil.copy(target, backup_path)
    
    # Update personalized backup log
    log_file = target.parent / f"backup_log_{student_id}.txt"
    with open(log_file, "a", encoding="utf-8") as log:
        log_entry = (
            f"Student: {student_id} ({student_name})\n"
            f"Original: {target.name}\n"
            f"Timestamp: {timestamp}\n"
            f"Size: {target.stat().st_size} bytes\n"
            f"Path: {backup_path}\n"
            f"{'-'*30}\n"
        )
        log.write(log_entry)
    
    print(f"Backup completed for {student_id} ({student_name})")
    print(f"File saved as: {backup_name}")

# Test run
test_file = activity_path / "Act5_example.txt"
smart_backup(test_file)
```
