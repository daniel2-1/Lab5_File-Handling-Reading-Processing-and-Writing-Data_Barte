```python
from datetime import datetime
import shutil
from pathlib import Path

# Step 2: Define working directory (Updated to include Surname)
backup_dir = Path.home() / "Documents" / "Barte_Activity_5"
backup_dir.mkdir(exist_ok=True)

# Task 2: Function to list only backups for a given filename
def list_backups_for_file(filename: str):
    stem = Path(filename).stem
    print(f"\n--- Backups for {filename} ---")
    # Search for files that start with the filename stem and include 'Barte_backup'
    for backup in backup_dir.glob(f"{stem}_Barte_backup*"):
        print("-", backup.name)

# Task 1 & 3: Modified write function with Surname and Overwrite/Append choice
def write_with_backup_v2(filename: str, content: str, mode: str = "w"):
    file_path = backup_dir / filename
    
    if file_path.exists():
        timestamp = datetime.now().strftime("%Y%m%d_%H%M%S")
        # Added Surname to the backup format [cite: 371]
        backup_name = f"{file_path.stem}_Barte_backup_{timestamp}{file_path.suffix}"
        backup_path = file_path.with_name(backup_name)
        shutil.copy2(file_path, backup_path)
        print(f"Backup saved: {backup_path.name}")
    
    # Task 3: Use the 'mode' parameter to let users choose behavior [cite: 373]
    with open(file_path, mode, encoding="utf-8") as file:
        file.write(content)
    
    status = "updated (appended)" if mode == "a" else "saved (overwritten)"
    print(f"File {status}: {file_path.name}")

# Demonstration
write_with_backup_v2("demo.txt", "\nMore content", mode="a")
list_backups_for_file("demo.txt")
```

```python
from pathlib import Path
from datetime import datetime
import shutil

# --- SETUP: WORKING DIRECTORY ---
# Folder name includes Surname
output_dir = Path.home() / "Documents" / "Barte_Activity_5"
output_dir.mkdir(exist_ok=True)
file_path = output_dir / "Act5_example.txt"

# --- PART 1: WRITING TO A FILE ---
with open(file_path, "w", encoding="utf-8") as file:
    file.write("Hello, Welcome to Python Programming!\n")
    file.write("File saved safely with pathlib.\n")
    file.write("Python makes file handling easy!\n")
print(f"Part 1: File saved to {file_path}")

# --- PART 2: READING FROM A FILE ---
print("\nPart 2: Filtered Reading (Lines with 'Python'):")
if file_path.exists():
    with open(file_path, "r", encoding="utf-8") as file:
        for line_number, line in enumerate(file, 1):
            clean_line = line.strip()
            if "Python" in clean_line:
                word_count = len(clean_line.split())
                print(f"Line {line_number} ({word_count} words): {clean_line}")

# --- PART 3: APPENDING DATA ---
extra_lines = ["\nFirst appended line.", "\nSecond appended line."]
with open(file_path, "a", encoding="utf-8") as file:
    file.writelines(extra_lines)
    user_note = input("\nPart 3: Enter text to append to the file: ")
    file.write(f"\nUser Note: {user_note}")

# --- PART 4: SAFE OPERATIONS & BACKUP ---
def list_backups_for_file(filename):
    stem = Path(filename).stem
    print(f"\n--- Backups for {filename} ---")
    for backup in output_dir.glob(f"{stem}_Barte_backup*"):
        print("-", backup.name)

def write_with_backup(filename, content):
    target_path = output_dir / filename
    if target_path.exists():
        timestamp = datetime.now().strftime("%Y%m%d_%H%M%S")
        b_name = f"{target_path.stem}_Barte_backup_{timestamp}{target_path.suffix}"
        shutil.copy2(target_path, output_dir / b_name)
    
    with open(target_path, "w", encoding="utf-8") as f:
        f.write(content)

# --- PART 5: MENU-DRIVEN FILE MANAGER ---
def file_manager():
    f_name = input("\nEnter filename to manage (e.g., notes.txt): ")
    f_path = output_dir / f_name
    
    while True:
        print("\n--- MENU ---")
        print("1. Write | 2. Append | 3. Read | 4. Backup | 5. Exit | 6. List All Backups")
        choice = input("Choose an option (1-6): ")

        if choice == "1":
            content = input("Content: ")
            with open(f_path, "w") as f: f.write(content)
        elif choice == "2":
            content = input("Append: ")
            with open(f_path, "a") as f: f.write("\n" + content)
        elif choice == "3":
            if f_path.exists(): 
                print("\n--- File Content ---")
                print(f_path.read_text())
            else: 
                print("Error: File not found.")
        elif choice == "4":
            text = f_path.read_text() if f_path.exists() else ""
            write_with_backup(f_name, text)
            print("Backup created successfully.")
        elif choice == "6":
            print("\nAll Backups in Directory:")
            for b in output_dir.glob("*_backup_*"): 
                print(f" - {b.name}")
        elif choice == "5":
            print("Exiting...")
            break

if __name__ == "__main__":
    file_manager()
```

```python
# Merge Two Files
from pathlib import Path

# Personal details
student_id = "2025-0419"

# Define file paths
documents_path = Path.home() / "Documents" /"Activity_5_Files"
f1 = documents_path / f"intro_{student_id}.txt"
f2 = documents_path / f"lines_{student_id}.txt"
merged = documents_path / f"merged_{student_id}.txt"

# Ensure both source files exist
if not f1.exists():
   f1.write_text(f"Welcome {student_id} to File Handling in Python!")
print(f"Sample intro file created for Student ID:{student_id}")
if not f2.exists():
    f2.write_text("Line 1\nLine 2\nLine 3")
print(f"Sample lines file created for Student ID:{student_id}")

# Merge contents of both files
with merged.open("w") as mf:
   mf.write(f1.read_text())
   mf.write("\n")
   mf.write(f2.read_text())
print(f"Files merged successfully for Student ID:{student_id}")
```
